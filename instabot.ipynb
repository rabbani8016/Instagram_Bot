```python
from instagrapi import Client

bot = Client()
bot.login("rabbani.gres","password")

# Example : Upload an image
# bot.photo_upload("Instabotpic.jpg","This is test file")

# Get followers of a user
user_id = bot.user_id_from_username("rabbani.gres")
followers = bot.user_followers(user_id)

for follower_id, follower_info in followers.items():
    print(f"{follower_info.username} - {follower_info.full_name}")
```
